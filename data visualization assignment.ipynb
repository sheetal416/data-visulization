```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X_reduced = PCA(n_components = 3).fit_transform(iris.data)
Y = iris.target
chart = plt.figure(1,figsize = (10,8))
axis = Axes3D(chart, elev = -110, azim = 100)
axis.scatter(X_reduced[:, 0],X_reduced[:, 1],X_reduced[:, 2], c = Y,cmap = plt.cm.rainbow_r)
axis.set_title("three different direction")
axis.set_xlabel("my first transform data")
axis.set_ylabel("my second transform data")
axis.set_zlabel("my third transform data")

plt.show()
```
